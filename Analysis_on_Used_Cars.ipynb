**Data Transformation**

'Price','mileage','engV' are highly skewed so just to make normalized I will be using log transforamtion which help in normalization ,so these variables can have standard scale with other variables.

```
#Function for the col name variables using log transformation
def log_transform(data_final_usedcar,col):
  for colname in col:
    if(data_final_usedcar[colname]==1).all:
      data_final_usedcar[colname + "_log"]=np.log(data_final_usedcar[colname]+1)
    else:
      data_final_usedcar[colname + "_log"]=np.log(data_final_usedcar[colname])    
log_transform(data_final_usedcar,['price','mileage','engV','year'])
plt.figure(figsize=(30,14))
plt.subplot(2,2,1)
sns.distplot(data_final_usedcar["mileage_log"],axlabel='mileage_log');
plt.subplot(2,2,2)
sns.distplot(data_final_usedcar["year_log"],axlabel='year_log');
plt.subplot(2,2,3)
sns.distplot(data_final_usedcar["price_log"],axlabel='price_log');
plt.subplot(2,2,4)
sns.distplot(data_final_usedcar['engV_log'],axlabel='engV_log');


plt.show
```
